```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("state_data.csv")
```

```python
df.shape
# 52 rows and 4 columns
```

```python
head = df.head()
tail = df.tail()
print(head)
# each column means state, latitude, lonitude, and its name
# the first state in the dataset is Alaska
# the last state in the dataset is Wyoming
# There are a total of 51 entries
# there are more than 50 states, because it included DC (district Columbia)
print(tail)
```

```python
med_lat = df['latitude'].median()
med_lon = df['longitude'].median()
print(f"Median latitude : {med_lat:.4f}")
print(f"Median longitude: {med_lon:.4f}")
```

```python
med_lat = df['latitude'].median()
df['abs_from_med_lat'] = (df['latitude'] - med_lat).abs()
```

```python
med_lon = df['longitude'].median()
df['abs_from_med_lon'] = (df['longitude'] - med_lon).abs()
```

```python
rows, cols = df.shape
print(f"Rows: {rows}, Columns: {cols}")
```

```python
df
```

```python
i_min_lat = df['abs_from_med_lat'].idxmin()
i_max_lat = df['abs_from_med_lat'].idxmax()
print(f"Closest to median latitude: {df.loc[i_min_lat, 'name']} by {df.loc[i_min_lat, 'abs_from_med_lat']:.2f}")
print(f"Farthest from median latitude: {df.loc[i_max_lat, 'name']} by {df.loc[i_max_lat, 'abs_from_med_lat']:.2f}")

# closest / farthest by longitude
i_min_lon = df['abs_from_med_lon'].idxmin()
i_max_lon = df['abs_from_med_lon'].idxmax()
print(f"Closest to median longitude: {df.loc[i_min_lon, 'name']} by {df.loc[i_min_lon, 'abs_from_med_lon']:.2f}")
print(f"Farthest from median longitude: {df.loc[i_max_lon, 'name']} by {df.loc[i_max_lon, 'abs_from_med_lon']:.2f}")
```

```python
close_lat = df.query('abs_from_med_lat < 1.0')[['name','abs_from_med_lat']]
print(close_lat.head())
# this is something I noticed that it returns rows of df that satisfy a boolean expression written as a string. It is used for 
# readable filters. 
```

```python
df = sns.load_dataset("diamonds")
print(df.shape, df.head())
df = df.dropna(axis=1)
print("Shape:", df.shape)
print("Rows:", df.shape[0], " Columns:", df.shape[1])
print(df.head())
print(df.tail())
# 53940 rows and 10 columns, each columns means the carat, its cut, the color, 
# its clarity, the depth, table, and the price, also x, y, z, meaning length, width and depth in mm. 
```

```python
sns.set(style="whitegrid")

fig, axes = plt.subplots(2, 2, figsize=(12, 9))

# Top-left: scatter (legend shown)
sns.scatterplot(data=df, x="carat", y="price", hue="cut",
                s=12, alpha=0.5, ax=axes[0, 0])
axes[0, 0].set_title("Carat vs. Price")
axes[0, 0].set_xlabel("carat (ct)")
axes[0, 0].set_ylabel("price (USD)")
axes[0, 0].legend(title="cut", fontsize=8)
# Each point is one diamond; color shows cut. Bigger carat generally means a higher price, 
# Cut groups partially separate, suggesting quality also matters.

# Top-right: different scatter
sns.scatterplot(data=df, x="depth", y="table", hue="color",
                s=12, alpha=0.4, palette="tab10", ax=axes[0, 1], legend=False)
axes[0, 1].set_title("Depth vs. Table")
axes[0, 1].set_xlabel("depth (%)")
axes[0, 1].set_ylabel("table (%)")
# Two geometry/quality metrics. 
# The cloud is wide with no strong linear relation meaning these features capture different aspects; there are a few outliers.

# Bottom-left: histogram
sns.histplot(df["price"], bins=50, ax=axes[1, 0], color="C2")
axes[1, 0].set_title("Distribution of Diamond Prices")
axes[1, 0].set_xlabel("price (USD)")
axes[1, 0].set_ylabel("count")
axes[1, 0].set_xscale("log")       # optional: log makes the skew easier to see
axes[1, 0].text(0.02, 0.95, "log x-scale", transform=axes[1, 0].transAxes, fontsize=8)
#Prices are right-skewed; most diamonds are relatively inexpensive with a long tail of high prices.

# Bottom-right:boxplot
sns.boxplot(data=df, x="cut", y="price", color="C3", ax=axes[1, 1])
axes[1, 1].set_title("Price by Cut")
axes[1, 1].set_xlabel("cut")
axes[1, 1].set_ylabel("price (USD)")
axes[1, 1].tick_params(axis="x", rotation=25)
# Boxplots compare price distributions across cut. 
# Higher cuts tend to have higher medians but theres large overlap, showing price also depends on other features.

plt.tight_layout()
plt.show()
```
