```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("exoplanet_data.csv", comment="#")
print("shape:", df.shape)
print("rows:", len(df), "  columns:", df.shape[1])
cols = df.columns.tolist()
print("\nColumn titles:")
for c in cols:
    print(c)
# I have no idea what decstr, dec, sy_dist, sy_disterr1 and sy_disterr2 means
```

```python
# Database Column Name: pl_name
# Description: Official designation of the exoplanet in the archive.
# Its the planet’s name.

# Database Column Name: hostname
# Description: Name of the star (or stellar system) the planet orbits.
# Its the parent star’s name.

# Database Column Name: discoverymethod
# Description: Primary technique used to discover the planet (e.g., Transit, Radial Velocity, Imaging, Microlensing, Timing).
# Basically the method how astronomers first detected it.

# Database Column Name: st_spectype
# Description: MK spectral classification of the host star, indicating temperature and luminosity class.
# Its the star’s color/temperature class.

# Database Column Name: pl_eqt
# Description: Estimated planetary equilibrium temperature in Kelvin from a simple energy-balance model
# Its a rough predicted average temperature.

# Database Column Name: pl_bmassj
# Description: “Best” planet mass in Jupiter masses; if only a radial-velocity minimum mass is known, this stores M·sin(i).
# It is the planet’s mass in units of Jupiter’s mass.
```

```python
wanted = ['pl_name', 'hostname', 'discoverymethod', 'st_spectype', 'pl_eqt', 'pl_bmassj']
keep = [c for c in wanted if c in df.columns]
small = df[keep].copy()
small_clean = small.dropna()
print('shape:', small_clean.shape)
rows, cols = small_clean.shape
print('rows:', rows, 'columns:', cols)
print(small_clean.head())
print(small_clean.tail())
print(list(small_clean.columns))
```

```python
use = small_clean.dropna(subset=['discoverymethod', 'pl_bmassj', 'pl_eqt'])
method_counts = use['discoverymethod'].value_counts()
print(method_counts)
print('Most common:', method_counts.idxmax(), method_counts.max())
print('Least common:', method_counts.idxmin(), method_counts.min())
# There are three different discovery methods, being transit, radial velocity and imaging. most being transit and least being imaging. 
# Transit: detect tiny dips in starlight when a planet crosses the star.
# Radial Velocity: measure the star’s wobble (Doppler shifts) due to a planet’s gravity.
# Direct Imaging: take pictures; works best for big, wide, bright/young planets.
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=use, x='pl_bmassj', y='pl_eqt', hue='discoverymethod')
plt.title('Planet Mass vs. Equilibrium Temperature')
plt.xlabel('Planet Mass [Jupiter masses]')
plt.ylabel('Equilibrium Temperature [K]')
plt.legend(title='Method', bbox_to_anchor=(1.02,1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=use, x='pl_bmassj', y='pl_eqt', hue='discoverymethod')
plt.xscale('log')
plt.yscale('log')
plt.title('Planet Mass vs. Equilibrium Temperature (log–log)')
plt.xlabel('Planet Mass [Jupiter masses] (log)')
plt.ylabel('Equilibrium Temperature [K] (log)')
plt.legend(title='Method', bbox_to_anchor=(1.02,1), loc='upper left')
plt.tight_layout()
plt.show()

# Direct imaging detections cluster at the high-mass end, so imaging is more likely to find massive planets and not small ones. 
```

```python
big = use[use['pl_bmassj'] > 5].dropna(subset=['pl_eqt'])
coldest = big.loc[big['pl_eqt'].idxmin()]

print('Coldest massive planet (>5 M_J):')
print('Name :', coldest['pl_name'])
print('Host :', coldest['hostname'])
print('Mass :', f"{coldest['pl_bmassj']:.2f} M_J")
print('Teq  :', f"{coldest['pl_eqt']:.0f} K")
print('Method:', coldest['discoverymethod'])
```
