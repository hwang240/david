```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def make_sine_wave(x, A, w):
    return A * np.sin(w * x)
    x = np.linspace(0, 2 * np.pi, 1000)
amplitude = [0.5, 1, 1.5, 2, 2.5]
frequency = [1, 2, 3, 4, 5]
fig, axes = plt.subplots(1, 3, figsize=(25, 5))
axes
for A in amplitude:
    y = make_sine_wave(x, A, 2)
    axes[0].plot(x, y, label=f"A={A}")

axes[0].set_title("Sine waves: varying amplitude (w=2)", fontsize=14)
axes[0].set_xlabel("x", fontsize=12)
axes[0].set_ylabel("y", fontsize=12)
axes[0].legend(fontsize=10)
axes[0].grid(True)

for w in frequency:
    y = make_sine_wave(x, 1, w)
    axes[1].plot(x, y, label=f"w={w}")

axes[1].set_title("Sine waves: varying frequency (A=1)", fontsize=14)
axes[1].set_xlabel("x", fontsize=12)
axes[1].set_ylabel("y", fontsize=12)
axes[1].legend(fontsize=10)
axes[1].grid(True)

for A, w in zip(amplitude, frequency):
    y = make_sine_wave(x, A, w)
    axes[2].plot(x, y, label=f"A={A}, w={w}")

axes[2].set_title("Sine waves: varying amplitude & frequency", fontsize=14)
axes[2].set_xlabel("x", fontsize=12)
axes[2].set_ylabel("y", fontsize=12)
axes[2].legend(fontsize=10)
axes[2].grid(True)
plt.show()
```
